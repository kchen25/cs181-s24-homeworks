```python
# standard imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import multivariate_normal as mvn
from scipy.stats import poisson
from scipy.stats import norm
```

## 2.1 Plot $y = -3x^2 + 72x + 70$.

```python
# Generate 50 evenly spaced points from -100 to 100
x = np.linspace(0, 25, 50)  

# Calculate corresponding y values 
y = -3 * x**2 + 72 * x + 70

# Create plot
plt.figure(figsize=(8, 6)) 
plt.plot(x, y, label='y = -3x^2 + 72x + 70', color='blue')
plt.title('Plot of y = -3x^2 + 72x + 70')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.show()
```

## 2.3 Plot $f(x_1, x_2)$.

```python
def function(x1, x2):
    return np.exp(-(x1-2)**2 - (x2-1)**2)

# Create grid
x1 = np.linspace(0, 6, 100)
x2 = np.linspace(0, 6, 100)
X1, X2 = np.meshgrid(x1, x2)  # Create grid of points in the 2D plane

Y = function(X1, X2) # vector of y-values for each point in the plane

# Create a 3D plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Use plot_surface to create the 3D surface plot
surf = ax.plot_surface(X1, X2, Y, cmap='viridis')  

# Add labels and color bar
ax.set_xlabel('X1')
ax.set_ylabel('X2')
ax.set_zlabel('Y')
fig.colorbar(surf, ax=ax, shrink=0.5, aspect=10)

# Show plot
plt.title('3D Surface Plot of y = f(x1, x2)')
plt.show()
```

## 3.1(a) Draw from joint distribution of $(S, W)$ and generate bivariate histogram.

```python
# Define mu and covariance matrix
mu = np.array([120, 4])
cov_matrix = np.array([[1.5, 1], [1, 1.5]])

# Sample 500 times from joint distribution
samples = mvn(mean=mu, cov=cov_matrix).rvs(size=500)

# Extract the size (S) and weight (W) samples
S_samples = samples[:, 0]
W_samples = samples[:, 1]
```

```python
# Create a bivariate histogram using seaborn
plt.figure(figsize=(8, 6))
sns.histplot(x=S_samples, y=W_samples, bins=20, cmap='viridis', cbar=True)
plt.xlabel('Size (in^3)')
plt.ylabel('Weight (lbs)')
plt.title('Bivariate Gaussian Distribution of Size and Weight')
plt.show()
```

## 3.2 Plot the joint PDFs at $S=118$ and $S=122$, varying $W$.

```python
# Define the fixed values of S
S_values = [118, 122]

# Generate evenly-spaced values of W
W_values = np.linspace(0, 10, 1001)

# Create subplots for both fixed S values
fig, axs = plt.subplots(len(S_values), 1, figsize=(8, 12), sharex=True)

# Calculate and plot the PDF for each fixed S value
for i, S in enumerate(S_values):
    pdf_values = mvn.pdf(x=np.vstack((np.ones(1001)*S, W_values)).T, mean=mu, cov=cov_matrix)
    
    axs[i].plot(W_values, pdf_values, label=f'S = {S}', color='blue')
    axs[i].set_title(f'Joint Bivariate Gaussian PDF for S = {S}')
    axs[i].set_ylabel('PDF')
    axs[i].grid(True)
    axs[i].legend()

# Add common x-axis label
plt.xlabel('W')
plt.tight_layout()
plt.show()
```

## 3.5(a) Function to simulate draws from $T^{*}$.

```python
def get_packages(num_packages):
    # For each package, generate S and W
    mu = np.array([120, 4])
    cov_matrix = np.array([[1.5, 1], [1, 1.5]])

    # Sample num_packages times from joint distribution to get (W, S)
    samples = mvn.rvs(mean=mu, cov=cov_matrix, size=num_packages)

    # Sample num_packages times from Norm(0,5) to get epsilon
    epsilon = norm(0, np.sqrt(5)).rvs(size=num_packages)

    # Calculate T for each package and sum
    T = 60 + 0.6*samples[:, 1] + 0.2*samples[:, 0] + epsilon
    return np.sum(T)

def sample_Tstar(size):
    # Number of packages is ditributed Pois(24 * 3) since it is a sum of 24 Pois r.v.s.
    num_packages = poisson.rvs(mu=24 * 3, size=size)
    vfunc = np.vectorize(get_packages)
    return vfunc(num_packages)
```

## 3.5(b) Empirically estimate the mean and standard deviation of $T^{*}$ by generating $1000$ samples.

```python
Tstar_samples = sample_Tstar(1000)
# Plotting a basic histogram
plt.hist(Tstar_samples, bins=30, color='skyblue', edgecolor='black')
 
# Adding labels and title
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Basic Histogram')
 
# Display the plot
plt.show()

print("T* sample mean: " + str(np.mean(Tstar_samples)))
print("T* sample std: " + str(np.std(Tstar_samples)))
```
